```python
#Continuous variable can take an "uncountable" number of values(say Length).
#Discrete variable can only take on a "countable" number of values. Discrete Random Variable can have an infinite number of values(eg.a set of natural numbers(1,2,3,...).
#In statistics a distribution of discrete variables is represented by PMF. And, a distribution of continuous variables represented by PDF.


#discrete random variable of rolling a dice
#PMF
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
die=stats.randint(1,7)
r=range(1,7)
prob=[die.pmf(i) for i in r]

prob1=die.pmf(r)

sample=np.random.randint(1,7,size=30)
plt.hist(sample,bins=30)

plt.style.use("seaborn-colorblind")
plt.bar(r,prob)

```

```python
#continous random variable
#PDF
import matplotlib.pyplot as plt
from numpy.random import uniform
uniform(low=-2,high=3,size=300)
s=uniform(low=-2,high=3,size=300)

plt.hist(s, bins=30, density=True)
plt.plot(s, [0.1]*len(s), 'o')
```
